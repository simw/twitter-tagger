```python
from matplotlib.ticker import FuncFormatter

def time_fmt(t, pos):
    return '{}:{:}'.format(int(t*24), int((t*24-math.floor(t*24))*60))
time_formatter = FuncFormatter(time_fmt)
```

```python
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday', 'Monday', 'Tuesday']

plot_data = pd.DataFrame()
plot_data['times'] = bin_midpoints
for i, histo in enumerate(histos):
    plot_data[day_names[i]] = histo[0]
    
plot_data.plot(x='times', figsize=(9,6))
plt.xlabel('Time (UTC)')
plt.gca().xaxis.set_major_formatter(time_formatter)
plt.gca().set_xticks([0, 1/6, 2/6, 3/6, 4/6, 5/6, 1])
plt.ylabel('Number per {} minutes'.format(bin_size*24*60))
plt.gca().set_title('Tweet number during the day')
```

How about trying to fit a sine curve to this?

```python
import math
from sklearn.linear_model import LinearRegression

fit_data = pd.DataFrame()
for i, histo in enumerate(histos):
    if day_names[i] in ['Saturday', 'Sunday']:
        continue
    tmp = pd.DataFrame()
    tmp['t'] = bin_midpoints
    tmp['y'] = histo[0]
    tmp = tmp[tmp['y'] != 0]
    fit_data = pd.concat([fit_data, tmp])

fit_data['sin(2pit)'] = fit_data['t'].apply(lambda t: math.sin(2*math.pi*t))
fit_data['cos(2pit)'] = fit_data['t'].apply(lambda t: math.cos(2*math.pi*t))

X = fit_data[['sin(2pit)', 'cos(2pit)']]
y = fit_data['y']

regr = LinearRegression()
regr.fit(X, y)
print('Coefficients: {}'.format(regr.coef_))
print('Intercept: {}'.format(regr.intercept_))
print('Variance score: {:.2%}'.format(regr.score(X, y)))
```

Remember, $\sin(t+\phi) = \sin t \cos \phi  + \cos t \sin \phi $

So, can solve $A \cos\phi = c1$ and $A \sin\phi = c2$, giving $A^2 = c1^2 + c2^2$ and $tan\phi = \frac{c2}{c1}$

```python
b = regr.intercept_
c1, c2 = regr.coef_

A = math.sqrt(c1**2 + c2**2)
phi = np.arctan(c2/c1)

print('b={}, A={}, phi={}'.format(b, A, phi))
```

```python
fn = lambda t: b + A * np.sin(2 * math.pi * t - phi)
# fn = lambda t: b + c1 * np.sin(2*math.pi*t) + c2 * np.cos(2*math.pi*t)
fn = np.vectorize(fn)
ys = fn(bin_midpoints)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
plot_data.plot(x='times', ax=ax1)
ax1.plot(bin_midpoints, ys, color='k', linewidth=2)
ax1.set_xlabel('Time (UTC)')
ax1.xaxis.set_major_formatter(time_formatter)
ax1.set_xticks([0, 1/6, 2/6, 3/6, 4/6, 5/6, 1])
ax1.set_ylabel('Number per {} minutes'.format(bin_size*24*60))

fit_data.plot(kind='scatter', x='t', y='y', ax=ax2)
ax2.plot(bin_midpoints, ys, color='k', linewidth=2)
ax2.set_xlim(0, 1)
ax2.set_ylim(0, 2000)
ax2.set_title('Tweet frequency excluding weekends')
ax2.set_xlabel('Time (UTC)')
ax2.xaxis.set_major_formatter(time_formatter)
ax2.set_xticks([0, 1/6, 2/6, 3/6, 4/6, 5/6, 1])
ax2.set_ylabel('Number per {} minutes'.format(bin_size*24*60))
```

(ok, I cheated and put in -phi in the above, I couldn't see where I'd flipped the sign)

```python
def fit_daily_histogram(pts_in, bin_width, ax):
    def is_weekend(dt):
        return '{0:%A}'.format(dt) in ['Saturday', 'Sunday']
    this_filter = [not(is_weekend(pt)) for pt in pts_in]
    
    pts = pts_in[this_filter]
    
    start_day = int(math.floor(date2num(pts_in.min())))
    end_day = int(math.ceil(date2num(pts_in.max())))
    pts = np.array([date2num(d) - start_day for d in pts])
    
    bins = np.arange(0, 1, bin_width)
    bin_midpoints = (bins + bin_width/2)[:-1]
    
    histos = [np.histogram(pts[(pts < day) & (pts > (day-1))]-(day-1), bins=bins) for day in range(1, (end_day-start_day))]
    freqs = pd.concat([pd.DataFrame(zip(h[0], h[1]), columns=['y', 't']) for h in histos])
    freqs = freqs[freqs['y'] != 0]
    
    freqs['sin(2pit)'] = freqs['t'].apply(lambda t: math.sin(2*math.pi*t))
    freqs['cos(2pit)'] = freqs['t'].apply(lambda t: math.cos(2*math.pi*t))
    
    X = freqs[['sin(2pit)', 'cos(2pit)']]
    y = freqs['y']
    
    regr = LinearRegression()
    regr.fit(X, y)
    
    b = regr.intercept_
    c1, c2 = regr.coef_

    A = math.sqrt(c1**2 + c2**2)
    phi = np.arctan(c2/c1)
    
    print('b={}, A={}, phi={}'.format(b, A, phi))
    
    fn = lambda t: b + A * np.sin(2 * math.pi * t - phi)
    # fn = lambda t: b + c1 * np.sin(2*math.pi*t) + c2 * np.cos(2*math.pi*t)
    fn = np.vectorize(fn)
    ys = fn(bin_midpoints)
    
    freqs.plot(kind='scatter', x='t', y='y', ax=ax)
    ax.plot(bin_midpoints, ys, color='k', linewidth=2)
    ax.set_xlim(0, 1)
    ax.set_title('Tweet frequency excluding weekends')
    ax.set_xlabel('Time (UTC)')
    ax.xaxis.set_major_formatter(time_formatter)
    ax.set_xticks([0, 1/6, 2/6, 3/6, 4/6, 5/6, 1])
    ax.set_ylabel('Number per {} minutes'.format(bin_size*24*60))
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 7))

ccy_data = data['created_at']
fit_daily_histogram(ccy_data, 0.025, ax1)

ccy = 'usdjpy'
ccy_data = data[data['keywords'] == "[u'" + ccy + "']"]['created_at']
fit_daily_histogram(ccy_data, 0.025, ax2)

ccy = 'usdmxn'
ccy_data = data[data['keywords'] == "[u'" + ccy + "']"]['created_at']
fit_daily_histogram(ccy_data, 0.1, ax3)
```

## A recommendation engine

Let's take the top 20 tweeters by followers, and 'like' them.
Then we'll take the top 20 tweeters by tweet count, and 'dislike' them.

Then we'll run this through a normal scikit-learn classifier system, and see if there are some other tweets that are predicted to be liked.

```python
like_users = tweeters.reset_index().sort('user_followers', ascending=False)['user'][:20]
dislike_users = tweeters.reset_index().sort('_id', ascending=False)['user'][:20]
all_users = tweeters.reset_index()['user']
unknown_users = all_users[~(all_users.isin(like_users) | all_users.isin(dislike_users))]

liked_tweets = data[data['user'].isin(like_users) & (data['lang'] == 'en')]
liked_tweets['like'] = 1
disliked_tweets = data[data['user'].isin(dislike_users)]
disliked_tweets['like'] = 0

likes_dislikes = pd.concat([liked_tweets, disliked_tweets])
likes_dislikes = likes_dislikes[['text', 'like']]

texts = likes_dislikes['text']
likes = likes_dislikes['like']

unknown_tweets = data[data['user'].isin(unknown_users)]
```

We'll just look at the text for this classifier (should probably also look at language ...)

```python
ngram_max = 1
binary = False
analyzer = 'word'

from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer(ngram_range=(1,ngram_max), binary=False, lowercase=True, analyzer=analyzer)
X = vect.fit_transform(texts)
y = likes

print(X.shape)
```

For now, just looking at LogisticRegression - we could look at the probabilities of liking under LogisticRegression, and then set a higher barrier than 50%.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import GaussianNB

clf = LogisticRegression()
# clf = LinearSVC()
# clf = GaussianNB() # Doesn't work on sparse matrix!

clf.fit(X, y)
```

```python
unknown_tweets['like'] = clf.predict(vect.transform(unknown_tweets['text']))
unknown_tweets['like'].sum() / unknown_tweets['like'].count()
```

```python
max_colwidth = pd.get_option('max_colwidth')
pd.set_option('max_colwidth', 85)
display(unknown_tweets[['user','text']][unknown_tweets['like'] == 1][:20])
pd.set_option('max_colwidth', max_colwidth)
```
